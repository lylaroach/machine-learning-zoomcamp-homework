```python
v = np.array([2, 4, 5, 6])
```

```python
u = np.array([1, 0, 0, 2])
u.shape
```

```python
def vector_vector_mult(u, v):
    assert u.shape[0] == v.shape[0]
    n = u.shape[0]
    result = 0.0
    for i in range(n):
        result = result + u[i] * v[i]
    return result
```

```python
vector_vector_mult(u, v)
```

```python
u.dot(v)
```

```python
U = np.array([[2, 4, 5 ,6],
              [1, 2, 1, 2],
              [3, 1, 2, 1]])
U
U.shape
```

```python
def matrix_vector_mult(U, v):
    assert U.shape[1] == v.shape[0]
    num_rows = U.shape[0]
    result = np.zeros(num_rows)
    for i in range(num_rows):
        result[i] = vector_vector_mult(U[i], v)
    return result
```

```python
matrix_vector_mult(U, v)
```

```python
U.dot(v)
```

```python
V = np.array([[1, 1, 2],
              [0, .5, 1],
              [0, 2, 1],
              [2, 1, 0]])
V
```

```python
def matrix_matrix_mult(U, V):
    assert U.shape[1] == V.shape[0]
    
    num_rows = U.shape[0]
    num_cols = V.shape[1]
    result = np.zeros((num_rows, num_cols))

    for i in range(num_cols):
        vi = V[:, i]
        Uvi = matrix_vector_mult(U, vi)
        result[:, i] = Uvi
    return result
```

```python
matrix_matrix_mult(U, V)
```

```python
U.dot(V)
```
